```python
import numpy as np
import sklearn
```

```python
n = np.random.randint(2, 6)
weights = []
bias = np.random.randint(-1000,1000)
x_train = []
y_train = []
x_test = []
y_test = []

def y(W,X,B,N):
    temp = 0
    for i in range(N):
        temp += W[i]*X[i]
    return temp + B

for i in range(n):
    weights.append(np.random.randint(-500,500))
for i in range(10000):
    temp = []
    for _ in range(n):
        temp.append(np.random.randint(-2000,2000))
    temp.append(np.random.randint(-100000,100000))
    x_train.append(temp)
    if x_train[i][n] <= y(weights, x_train[i][:n],bias,n):
        y_train.append(0)
    elif x_train[i][n] > y(weights,x_train[i][:n],bias,n):
        y_train.append(1)
    else:
        print('There was an error.')
    
for i in range(5000):
    temp = []
    for _ in range(n):
        temp.append(np.random.randint(-2000,2000))
    temp.append(np.random.randint(-10000,10000))
    x_test.append(temp)
    if x_test[i][1] <= y(weights,x_test[i],bias,n):
        y_test.append(0)
    elif x_test[i][1] > y(weights,x_test[i],bias,n):
        y_test.append(1)
    else:
        print('There was an error.')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)

from sklearn.svm import SVC
svc = SVC(kernel='linear', random_state=0,probability=True)
svc.fit(x_train,y_train)
```

```python
if n == 2: 
    print('Model: y = {0}X1  + {1}X2 + {2}'.format(weights[0],weights[1], bias)) 
    
elif n ==3: 
        print('Model: y = {0}X1 + {1}X2 + {2}X3 + {3}'.format(weights[0],weights[1],weights[2], bias)) 

elif n == 4: 
    print('Model: y = {0}X1 + {1}X2 + {2}X3 + {3}X4 + {4}'.format(weights[0],weights[1], weights[2],weights[3], bias))

elif n == 5: 
    print('Model: y = {0}X1 + {1}X2 + {2}X3 + {3}X4 + {4}X5 + {5}'.format(weights[0],weights[1],weights[2],weights[3],weights[4], bias)) 

else: 
    print('An error has occurred')
```

```python
fails = []
for i in range(len(x_test)):
    if y_test[i] == svc.predict(scaler.transform([x_test[i]])):
        print('Successs')
        print('Input: ', x_test[i])
        print('True answer: ', y_test[i])
        print('Prediction: ', svc.predict(scaler.transform([x_test[i]]))[0])
        print('-------------------------------------------------')
    else:
        print('Failure')
        print('Input: ', x_test[i])
        print('True answer: ', y_test[i])
        print('Prediction: ', svc.predict(scaler.transform([x_test[i]]))[0])
        fails.append([x_test[i],y_test[i],svc.predict(scaler.transform([x_test[i]]))[0]])
        print('-------------------------------------------------')
```

```python
score = svc.score(scaler.transform(x_test),y_test)
print('Accuray: ', score)
print('Total number of tests: ',len(x_test))
print('Fails: ', int(len(x_test)-score*len(x_test)))
print('')
print('')
for i in fails:
    print('Failure')
    print('Input: ',i[0])
    print('True answer: ', i[1])
    print('Prediction: ', i[2])
    print('-------------------------------------------------')
```
