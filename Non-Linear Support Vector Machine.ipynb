```python
import numpy as np
import sklearn
```

```python
x_train = []
y_train = []
x_test = []
y_test = []

def y(x):
    return -0.000009*(x-1000)**3 + 4500   # curve that separates into two classes

for i in range(50000):
    x_train.append([np.random.randint(0, 2001), np.random.randint(0, 10001)])
    if x_train[i][1] <= y(x_train[i][0]):
        y_train.append(0)
    elif x_train[i][1] > y(x_train[i][0]):
        y_train.append(1)
    else:
        print('There was an error.')
        
for i in range(5000):
    x_test.append([np.random.randint(0, 2001), np.random.randint(0, 10001)])
    if x_test[i][1] <= y(x_test[i][0]):
        y_test.append(0)
    elif x_test[i][1] > y(x_test[i][0]):
        y_test.append(1)
    else:
        print('There was an error.')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)

from sklearn.svm import SVC
svc = SVC(kernel='poly', random_state=0,probability=True, degree = 3, gamma = 'auto')
svc.fit(x_train,y_train)
```

```python
fails = []
for i in range(len(x_test)):
    if y_test[i] == svc.predict(scaler.transform([x_test[i]])):
        print('Successs')
        print('Input: ', x_test[i])
        print('True answer: ', y_test[i])
        print('Prediction: ', svc.predict(scaler.transform([x_test[i]]))[0])
        print('-------------------------------------------------')
    else:
        print('Failure')
        print('Input: ', x_test[i])
        print('True answer: ', y_test[i])
        print('Prediction: ', svc.predict(scaler.transform([x_test[i]]))[0])
        fails.append([x_test[i],y_test[i],svc.predict(scaler.transform([x_test[i]]))[0]])
        print('-------------------------------------------------')
```

```python
score = svc.score(scaler.transform(x_test),y_test)
print('Accuracy: ', score)
print('Total number of tests: ',len(x_test))
print('Fails: ', int(len(x_test)-score*len(x_test)))
print('')
print('')
for i in fails:
    print('Failure')
    print('Input: ',i[0])
    print('True answer: ', i[1])
    print('Prediction: ', i[2])
    print('-------------------------------------------------')
```

```python
arr = [1507,2700]   # <--- manual input

temp = 0
if arr[1] <= y(arr[0]):
    temp = 0
elif arr[1] > y(arr[0]):
    temp = 1
    
arr = scaler.transform([arr])  
if svc.predict(arr)[0] == temp:
    print('Success')
else:
    print('Failure')
    print('Answer: ', temp)
    print('Prediction: ', svc.predict(arr)[0])
```
